```python
import numpy as np
```

```python
def f(x):
    return 7*x**3 - 2*x**2 - 7*x + 8

def g(x):
    return 21*x**2 - 4*x - 7

iter = 0 
```

Bisection
Konsep : ada 2 titik A dan B, dimana kedua titiknya positif dan negatif, maka ada titik C yang nilainya 0

Steps :
1. Tentukan nilai A dan B
2. Mencari nilai C
3. Cek nilai C
4. Jika nilai C nya lebih kecil dari epsilonnya, maka root di temukan, jika belum maka akan mengupdate nilai A dan B



```python
def bisection(a, b, tolerance = 0.062):
    global iter
    if np.sign(f(a)) == np.sign(f(b)):
        print("Invalid Input")
        return
    c = (a+b)/2
    if np.abs(f(c)) < tolerance:
        print(f"Root Found At {c}")
        iter = 0
        return
    iter += 1
    print(f"{iter} Iteration: {c}")
    if np.sign(f(c)) == np.sign(f(a)):
        bisection(c, b)
    elif np.sign(f(c)) == np.sign(f(b)):
        bisection(a, c)
        
bisection(-6, 9) 
```

Newton Raphson : mencari root dari garis singgung dengan cara 

Rumus :
 x1 = x0 - f(x0) / g(x0)

Steps :
1. Tentukan nilai x0
2. Cari root dari garis singung x0
3. Cek apakah x1 merupakan root atau tidak
4. Jika iya, root ketemu, jika belum maka mengupdate nilai x0 menjadi x1



```python
def newtonRhapson(x0, tolerance = 0.062):
    global iter
    iter += 1
    x1 = x0 - (f(x0) / g(x0)) 
    if np.abs(f(x1)) < tolerance:
        print(f"Root Found At {x1}")
        iter = 0
        return
    print(f"{iter} Iteration: {x1}")
    newtonRhapson(x1)
    
newtonRhapson(69)
```
